```python
# %% 【真の最終完成版】最適化ループ
import jax
import jax.numpy as jnp
from functools import partial
import matplotlib.pyplot as plt # 可視化のためにインポート

# 既存のコンポーネントをインポート
from src.mera.network import MERANetwork, MERAParams
from src.mera.hamiltonians import construct_3site_ising_hamiltonian
from src.mera.expectation import ExpectationValueCalculator
from src.mera.optimization import create_optimizer, OptimizerState

# --- ステップ1: パラメータ設定 ---
NUM_STEPS = 100
LEARNING_RATE = 0.005 # 学習率を少し小さくする

print("\n" + "="*60)
print("FINAL EXPERIMENT: Full Optimization Loop")
print(f"Running for {NUM_STEPS} steps with learning rate {LEARNING_RATE}")
print("="*60)

# --- ステップ2: セットアップ ---
network = MERANetwork(num_sites=9, bond_dim=4)
key = jax.random.PRNGKey(43) # 対称性を破るため、42以外のシードを使用
params = network.initialize(key)
hamiltonian = construct_3site_ising_hamiltonian(h_field=1.0)
calculator = ExpectationValueCalculator(mera_network=network)
optimizer = create_optimizer(method="adam", learning_rate=LEARNING_RATE)

print("✓ All components initialized successfully.")

# --- ステップ3: エネルギー関数と勾配関数の定義 ---
@partial(jax.jit, static_argnames=['calc'])
def energy_function(params, hamiltonian, calc):
    return calc.compute_expectation_value(hamiltonian, params)

energy_and_grad_fn = jax.value_and_grad(energy_function)

# --- ステップ4: 最適化ループの実行 ---
state = OptimizerState(iteration=0, params=params)

# 履歴を保存するための空リストを正しく初期化
energy_history = []  # ← ここを修正
grad_norm_history = []  # ← ここを修正

for step in range(NUM_STEPS):
    # エネルギーと勾配を計算
    energy, grads = energy_and_grad_fn(state.params, hamiltonian, calculator)
    
    # 履歴を保存
    energy_history.append(energy)
    grad_norm = optimizer.compute_gradient_norm(grads)
    grad_norm_history.append(grad_norm)
    
    # 5ステップごとに進捗を出力
    if step % 5 == 0:
        print(f"Step {step:3d}/{NUM_STEPS} | Energy: {energy:.6f} | Grad Norm: {grad_norm:.6f}")
        
    # オプティマイザでパラメータを更新
    state = optimizer.adam_step(state, grads)

print(f"Step {NUM_STEPS:3d}/{NUM_STEPS} | Energy: {energy_history[-1]:.6f} | Grad Norm: {grad_norm_history[-1]:.6f}")
print("\n✅ Optimization loop completed!")

# --- ステップ5: 結果の可視化 ---
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# エネルギーの履歴をプロット
ax1.plot(energy_history, marker='o', linestyle='-', label='Energy')
ax1.set_ylabel("Energy")
ax1.set_title("Optimization History")
ax1.grid(True)
ax1.legend()

# 勾配ノルムの履歴をプロット
ax2.plot(grad_norm_history, marker='o', linestyle='-', color='r', label='Gradient Norm')
ax2.set_ylabel("Gradient Norm")
ax2.set_xlabel("Optimization Step")
ax2.set_yscale('log') # 対数スケールで表示
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import os
import shutil
from pathlib import Path

def cleanup_project():
    """プロジェクトのクリーンアップ（JAX環境対応）"""
    
    # 現在のディレクトリから開始
    project_root = Path(".")
    
    # 削除したファイル/ディレクトリのカウント
    removed_count = {
        "__pycache__": 0,
        "*.pyc": 0,
        ".DS_Store": 0
    }
    
    # __pycache__ディレクトリを削除
    for pycache in project_root.rglob("__pycache__"):
        try:
            shutil.rmtree(pycache)
            removed_count["__pycache__"] += 1
        except Exception as e:
            print(f"Warning: Could not remove {pycache}: {e}")
    
    # .pycファイルを削除
    for pyc_file in project_root.rglob("*.pyc"):
        try:
            pyc_file.unlink()
            removed_count["*.pyc"] += 1
        except Exception as e:
            print(f"Warning: Could not remove {pyc_file}: {e}")
    
    # .DS_Storeファイルを削除（macOSの場合）
    for ds_store in project_root.rglob(".DS_Store"):
        try:
            ds_store.unlink()
            removed_count[".DS_Store"] += 1
        except Exception as e:
            print(f"Warning: Could not remove {ds_store}: {e}")
    
    # 結果を表示
    print("🧹 Cleanup Summary:")
    for file_type, count in removed_count.items():
        if count > 0:
            print(f"  - Removed {count} {file_type}")
    
    print("\n✅ Cleanup complete!")

# 実行
cleanup_project()
```
